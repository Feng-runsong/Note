```python
import numpy as np
```

```python
arr = np.arange(0, 10)
```

```python
# 最值
print(arr.min())
print(arr.max())

# 中位数
print(np.median(arr))

# 四分位点
Q1 = np.percentile(arr, 25)
Q1 = np.percentile(arr, 50)
Q1 = np.percentile(arr, 75)

# 求和/均值
print(arr.sum())
print(arr.mean())

# 标准差/方差
print(arr.std())
print(arr.var())
```
